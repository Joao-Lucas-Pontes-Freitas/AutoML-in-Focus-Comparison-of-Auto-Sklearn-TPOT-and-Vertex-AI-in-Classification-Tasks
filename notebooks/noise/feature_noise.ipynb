# Feature Noise

```python
import sys
DIR_ROOT = '../../'
sys.path.append(DIR_ROOT)
from src.noise import *
import pandas as pd

# Seleciona o dataset
DATASET_NAME = 'imdb'
df = pd.read_csv(f'{DIR_ROOT}data/{DATASET_NAME}_train.csv')

# Copia os dados originais e separa
X_train = df.drop(columns=['target'])
y_train = df['target'].to_frame()
```

## 10% Feature Noise - Tabular

```python
#Define the categorical columns
colunas_categoricas = []

colunas_numericas = X_train.columns.difference(colunas_categoricas).tolist()

X_train_noisy = data_noise_tabular(X_train, y_train, numerical_columns=colunas_numericas, categorical_columns=colunas_categoricas, p=0.1)

df_compare_data = X_train - X_train_noisy
df_compare_data = df_compare_data[(df_compare_data != 0).any(axis=1)]
df_compare_data
```

## 10% Feature Noise - Images

```python
X_train_noisy = data_noise_images(X_train, y_train, DATASET_NAME, p=0.1)

df_compare_data = X_train - X_train_noisy
df_compare_data = df_compare_data[(df_compare_data != 0).any(axis=1)]
df_compare_data
```

## 10% Feature Noise - Text

```python
X_train_noisy = data_noise_text(X_train, y_train, col_name='review', p=0.1)
df['review_noisy'] = X_train_noisy['review']
df_compara_data = df[df['review'] != df['review_noisy']]
df_compara_data[['review', 'review_noisy']]
df_compara_data
```

## Saving the Results

```python
# Gera o novo df com a combinação dos ruídos e salva na pasta certa.
new_df = X_train_noisy.copy()
new_df['target'] = y_train

new_df.to_csv(f'{DIR_ROOT}data/noise/feature/{DATASET_NAME}_train.csv', index=False)

# Mantém os dados de teste intactos e copia para a pasta nova
df_test = pd.read_csv(f'{DIR_ROOT}data/{DATASET_NAME}_test.csv')
df_test.to_csv(f'{DIR_ROOT}data/noise/feature/{DATASET_NAME}_test.csv', index=False)
```
